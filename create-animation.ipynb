## Gerando um gif com ImageIO

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imageio
```

```python
flag = cv2.imread('flag.png')

flag = cv2.cvtColor(flag, cv2.COLOR_BGR2RGB)
print(flag.shape)
plt.imshow(flag)
```

```python
flags = [i for i in range(10)]

flags[0] = cv2.imread('flag_0.png')
flags[1] = cv2.imread('flag_142.png')
flags[2] = cv2.imread('flag_284.png')
flags[3] = cv2.imread('flag_426.png')
flags[4] = cv2.imread('flag_568.png')
flags[5] = cv2.imread('flag_711.png')
flags[6] = cv2.imread('flag_853.png')
flags[7] = cv2.imread('flag_995.png')
flags[8] = cv2.imread('flag_1137.png')
flags[9] = cv2.imread('flag_1280.png')

for i in range(10):
    flags[i] = cv2.cvtColor(flags[i], cv2.COLOR_BGR2RGB)

flag_stack = np.vstack((np.hstack(flags[:5]), np.hstack(flags[5:])))

plt.figure(figsize=(10,10))    
plt.imshow(flag_stack)
```

```python
with imageio.get_writer("flag.gif", mode="I") as writer:
    for idx, frame in enumerate(flags):
        writer.append_data(frame)
```

## Gerando um vdeo com OpenCV

```python
from cv2 import VideoWriter, VideoWriter_fourcc
```

```python
width = flag.shape[1]
height = flag.shape[0]
FPS = 5
seconds = 2

fourcc = VideoWriter_fourcc(*'MP42')
video = VideoWriter('./flag-video.mp4', fourcc, float(FPS), (width, height))

for i in range(FPS*seconds):
    flags[i] = cv2.cvtColor(flags[i], cv2.COLOR_BGR2RGB)
    frame = flags[i]
    video.write(frame)
    
video.release()
```
